# L'importation des librairies

```python
# Pour la convertion du text au son
from gtts import gTTS

# Pour l'acceleration de l'audio
from pydub import AudioSegment

#Pour l'affichage de l'audio dans le notebook
from IPython.display import Audio 
```

# Notre premier example

```python
mytext = "Hello !"

language = 'en'

myobj = gTTS(text=mytext, lang=language, slow=False)

myobj.save("welcome.mp3")

display(Audio(filename="welcome.mp3"))
```

# La consultation des langues

```python
import gtts.lang

# Access the dictionary of supported languages
languages = gtts.lang.tts_langs()

# Print all languages and codes
for code, name in languages.items():
  print(f"{code}: {name}")
```

# Essayer le parametre de slow

```python
# Your text content
text_to_read = "C'est un example de lecture a des vitesses differentes ."

# Define language (optional, defaults to 'en')
language = 'fr'

# Vitesse de lecture lente
slow_speech = gTTS(text=text_to_read, lang=language, slow=True)

#Vitesse de lecture rapide
fast_speech = gTTS(text=text_to_read, lang=language, slow=False)


# Save the audio files (optional)
slow_speech.save("slow_reading.wav")
fast_speech.save("fast_reading.wav")
```
